## Insights for Tip Dataset

- The dataset contains 244 rows and 7 columns related to restaurant bills and tipping behavior.
- Numerical features: `total_bill`, `tip`, and `size`  
  **Categorical features: `sex`, `smoker`, `day`, and `time`.
- Most records are from Dinner time, especially on weekends (Saturday and Sunday).
- Male customers appear more frequently than female customers.
- The majority of customers are non-smokers.
- Tips generally increase with the total bill amount, indicating a positive correlation.
- Most dining groups (`size`) range between 2 and 3 people.
- The highest bills and tips are mostly observed during Saturday dinner.

## Insights from Descriptive Statistics

- The average total bill is approximately 19.79, while the average tip is around 3.00.
- Bills show high variability (standard deviation ≈ 8.90), indicating a wide range of spending.
- Most tips fall between 2.00 and 3.56, with a maximum tip of 10.00.
- The median bill (17.80) and **median tip (2.90) suggest a slightly right-skewed distribution.
- Most dining groups consist of 2 people, with group sizes ranging from 1 to 6.
- Larger bills and higher tips are less common but significantly influence the maximum values.
